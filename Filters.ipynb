# 1.0 Setup

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('dummy.jpg')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

# 2.0 Utility Functions

```python
def increaseIntensity(image, colorChannel, amount):
    (b, g, r) = cv2.split(image)
    if colorChannel == 'r':
        r[r > (255 - amount)] = (255 - amount)
        r += amount
    if colorChannel == 'g':
        g[g > (255 - amount)] = (255 - amount)
        g += amount
    if colorChannel == 'b':
        b[b > (255 - amount)] = (255 - amount)
        b += amount
    return cv2.merge((b, g, r))

def black_mask_overlay(image,alpha = 0.3):
    img = image.copy()
    overlay = img.copy()
    cv2.rectangle(overlay, (0, 0), (2000, 2000),(0, 0, 0), -1)
    cv2.addWeighted(overlay, alpha, img, 1 - alpha,0, img)
    return img
```

# 3.0 Filters

## 3.1 Charming Filter

```python
img = image.copy()
img = increaseIntensity(img, 'r', 60)
img = increaseIntensity(img, 'b', 5)
img = black_mask_overlay(img, 0.15)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## 3.2 Gray Filter

```python
img = image.copy()
plt.imshow(cv2.cvtColor(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY),cv2.COLOR_GRAY2RGB))
plt.show()
```

## 3.3 SummerTime Filter

```python
img = image.copy()
img = increaseIntensity(img, 'r', 10)
img = increaseIntensity(img, 'g', 12)
img = black_mask_overlay(img, 0.1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## 3.4 1997 Filter

```python
img = image.copy()
img = increaseIntensity(img, 'r', 40)
img = increaseIntensity(img, 'b', 25)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## 3.5 Gotham Filter

```python
img = image.copy()
img = increaseIntensity(img, 'b', 50)
img = black_mask_overlay(img, 0.1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
